# Start of Outliers Handling

```python
sns.boxplot(x='Age', data=temp_df)
plt.title('Boxplot of Age')
plt.show()
```

```python
Q1 = temp_df['Age'].quantile(0.25)
Q3 = temp_df['Age'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
cut_off = IQR * 1.5
lower = Q1 - cut_off
upper =  Q3 + cut_off
print(lower,upper)
df1 = temp_df[temp_df['Age']> upper]
df2 = temp_df[temp_df['Age'] < lower]
print('Total number of outliers are', df1.shape[0]+ df2.shape[0])
df_After_Age = temp_df[(temp_df['Age'] < upper) & (temp_df['Age'] > lower)]

print(temp_df.shape)
print(df_After_Age.shape)
```

```python
sns.boxplot(x='Age', data=df_After_Age)
plt.title('Boxplot of Age')
plt.show()
```

```python
sns.boxplot(x='Height', data=df_After_Age)
plt.title('Boxplot of Height')
plt.show()
```

```python
Q1 = df_After_Age['Height'].quantile(0.25)
Q3 = df_After_Age['Height'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
cut_off = IQR * 1.5
lower = Q1 - cut_off
upper =  Q3 + cut_off
print(lower,upper)
df1 = df_After_Age[df_After_Age['Height']> upper]
df2 = df_After_Age[df_After_Age['Height'] < lower]
print('Total number of outliers are', df1.shape[0]+ df2.shape[0])
df_After_Height = df_After_Age[(df_After_Age['Height'] < upper) & (df_After_Age['Height'] > lower)]

print(df_After_Age.shape)
print(df_After_Height.shape)
```

```python
sns.boxplot(x='Height', data=df_After_Height)
plt.title('Boxplot of Height')
plt.show()
```

```python
sns.boxplot(x='Year', data=df_After_Height)
plt.title('Boxplot of Year')
plt.show()
```

```python
Q1 = df_After_Height['Year'].quantile(0.25)
Q3 = df_After_Height['Year'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
cut_off = IQR * 1.5
lower = Q1 - cut_off
upper =  Q3 + cut_off
print(lower,upper)
df1 = df_After_Height[df_After_Height['Year']> upper]
df2 = df_After_Height[df_After_Height['Year'] < lower]
print('Total number of outliers are', df1.shape[0]+ df2.shape[0])
df_After_Year = df_After_Height[(df_After_Height['Year'] < upper) & (df_After_Height['Year'] > lower)]

print(df_After_Height.shape)
print(df_After_Year.shape)
```

```python
sns.boxplot(x='Year', data=df_After_Year)
plt.title('Boxplot of Year')
plt.show()
```

```python
sns.distplot(df_After_Year['Age'])
```

```python
from scipy import stats
import numpy as np

z_Age = np.abs(stats.zscore(df_After_Year['Age']))
z_Age
```

```python
filtered_entries = z_Age < 2.5
df_zscore_filter_A_zAge = df_After_Year[filtered_entries]
print(df_After_Year.shape)
print(df_zscore_filter_A_zAge.shape)
```

```python
sns.distplot(df_zscore_filter_A_zAge['Age'])
```

```python
sns.distplot(df_zscore_filter_A_zAge['Height'])
```

```python
z_Height = np.abs(stats.zscore(df_zscore_filter_A_zAge['Height']))
filtered_entries = z_Height < 2.4
df_zscore_filter_A_zHeight = df_zscore_filter_A_zAge[filtered_entries]
print(df_zscore_filter_A_zAge.shape)
print(df_zscore_filter_A_zHeight.shape)
```

```python
sns.distplot(df_zscore_filter_A_zHeight['Height'])
```

```python
sns.distplot(df_zscore_filter_A_zHeight['Weight'])
```

```python
z_Weight = np.abs(stats.zscore(df_zscore_filter_A_zHeight['Weight']))
filtered_entries = z_Weight < 3
df_zscore_filter_A_zWeight = df_zscore_filter_A_zHeight[filtered_entries]
print(df_zscore_filter_A_zHeight.shape)
print(df_zscore_filter_A_zWeight.shape)
```

```python
sns.distplot(df_zscore_filter_A_zWeight['Weight'])
```

```python
sns.distplot(df_zscore_filter_A_zWeight['Year'])
```

```python
fig_4, axs_5 = plt.subplots(1,1)
plt.xticks(rotation=90)
axs_5.set_xlabel("Sport'")
axs_5.set_ylabel('Number Of Medals')
axs_5.bar(df_Egypt.index,df_Egypt["Medal"]) # bottom: The y coordinate(s) of the bars bases (default: 0)
```

```python
df_Egypt_male= df_athelete_events[(df_athelete_events['Team']=='Egypt') & (df_athelete_events['Sex']=='M')].sort_values("Year").groupby("Sport").count()
df_Egypt_male = df_Egypt_male[(df_Egypt_male['Medal']>0)]
df_Egypt_male
```

```python
df_Egypt_female= df_athelete_events[(df_athelete_events['Team']=='Egypt') & (df_athelete_events['Sex']=='F')].sort_values("Year").groupby("Sport").count()
df_Egypt_female = df_Egypt_female[(df_Egypt_female['Medal']>0)]
df_Egypt_female
```

```python
for index, row in df_Egypt_male.iterrows():
    if(not index in (np.array(df_Egypt_female.index))):
        entry = df_Egypt_male.loc[index]
        df_Egypt_female = df_Egypt_female.append(entry)
        df_Egypt_female.loc[index,"Medal"] = 0

df_Egypt_female = df_Egypt_female.reindex(index=df_Egypt_male.index)
df_Egypt_female = df_Egypt_female.reset_index()
df_Egypt_male 
```

```python
fig, axs = plt.subplots(1,1)
plt.xticks(rotation=90)
axs.set_xlabel("Sport")
axs.set_ylabel('Number Of Medals')
axs.bar(df_Egypt_male.index,df_Egypt_male["Medal"],width=0.2,label="male",bottom=df_Egypt_female["Medal"])
axs.bar(df_Egypt_female.index,df_Egypt_female["Medal"],width=0.2,label="female")
axs.legend(loc=2)
```
