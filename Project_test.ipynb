## - Research quistion :- Which countries dominates the olympics over the years? 
### To answer this quistion we will get the accumilative count of the medals won by each country over the years and will consider the top 5 countris to be the dominating countries, Also we will try to differentiate between the medals won in the summer and winter olympics.

```python
# get the accumulative count of medals over all the OLympics
accumulativeResults=df_All_Olympics.groupby(['region'])['Bronze','Silver','Gold','Total_Medals'].sum().fillna(0).astype(int)
accumulativeResults=accumulativeResults.sort_values(by='Total_Medals', ascending=False)
accumulativeResults=accumulativeResults.reset_index()
accumulativeResults #accumulative count of medals over all the OLympics for each country
```

```python
# get the Summer/Winter accumulative count of medals over all the OLympics for the top 5 countries
accumulativeResults=df_All_Olympics.groupby(['region','Season'])['Bronze','Silver','Gold','Total_Medals'].sum().fillna(0).astype(int)
accumulativeResults=accumulativeResults.sort_values(by='Total_Medals', ascending=False)
accumulativeResults=accumulativeResults.reset_index()

#************************************
t=accumulativeResults.region.values
c= np.logical_or(t=="USA",t=="Russia")
c= np.logical_or(c,t=="Germany")
c= np.logical_or(c,t=="UK")
c= np.logical_or(c,t=="France")
topFive=accumulativeResults[c]
topFive #accumulative count of medals over all the OLympics for the top 5 countries for both Summer/Winter
topFive
```

```python
#helper Step to maintain the order
temp=topFive.query("Season=='Winter'")
temp['help']=[2,1,3,5,4]
temp=temp.sort_values(by='help')
#--------------------------
# plot bars in stack manner
plt.bar(topFive.region[topFive.Season=="Summer"].values, topFive.Total_Medals[topFive.Season=="Summer"].values, color='r')
plt.bar(temp.region.values, temp.Total_Medals.values,bottom=topFive.Total_Medals[topFive.Season=="Summer"].values,color='b')
# add legend
top_bar = mpatches.Patch(color='r', label='Summer')
bottom_bar = mpatches.Patch(color='b', label='Winter')
plt.legend(handles=[top_bar, bottom_bar])

plt.xlabel("Top 5 countries")
plt.ylabel("Number of Medals")
plt.title('Total Medals per Country')
# show the graph
plt.show()
```

```python
# Function to map country to city

def host_country(col):
    if col == "Rio de Janeiro":
        return "Brazil"
    elif col == "London":
        return "UK"
    elif col == "Beijing":
        return  "China"
    elif col == "Athina":
        return  "Greece"
    elif col == "Sydney" or col == "Melbourne":
        return  "Australia"
    elif col == "Atlanta" or col == "Los Angeles" or col == "St. Louis" or col=="Salt Lake City" or col=="Lake Placid" or col=="Squaw Valley":
        return  "USA"
    elif col == "Barcelona":
        return  "Spain"
    elif col == "Seoul":
        return  "South Korea"
    elif col == "Moskva" or col=="Sochi":
        return  "Russia"
    elif col == "Montreal" or col=="Calgary" or col =="Vancouver":
        return  "Canada"
    elif col == "Munich" or col == "Berlin" or col=="Garmisch-Partenkirchen":
        return  "Germany"
    elif col == "Mexico City":
        return  "Mexico"
    elif col == "Tokyo" or col=="Nagano"or col=="Sapporo" :
        return  "Japan"
    elif col == "Roma" or col=="Torino" or col=="Cortina d'Ampezzo":
        return  "Italy"
    elif col == "Paris" or col=="Albertville" or col=="Grenoble" or col=="Chamonix":
        return  "France"
    elif col == "Helsinki":
        return  "Finland"
    elif col == "Amsterdam":
        return  "Netherlands"
    elif col == "Antwerpen":
        return  "Belgium"
    elif col == "Stockholm":
        return  "Sweden"
    elif col == "Lillehammer" or col=="Oslo":
        return  "Norway"
    elif col == "Innsbruck":
        return  "Austria"
    elif col == "Sarajevo":
        return  "Bosnia and Herzegovina"
    elif col == "Sankt Moritz":
        return  "Switzerland"
    else:
        return "Other"


```

## The second research quistion is :- 
## Is there a correlation between height and weight and achieving medals in some sports such as basketball, football and swimming?
### To answer this quistion we will add a column rebresenting the BMI  to the athelets DataSet, Body Mass Index (BMI) is a person's weight in kilograms divided by the square of height in meters. A high BMI can indicate high body fatness. This column will help us to answer our research quistion.

```python
# ADD the BMI column
df_athelete_events_BMI = df_athelete_events.copy()
#------------------------------------------------------
# df_athelete_events_BMI["Age"].fillna(value=round(df_athelete_events["Age"].mean()), inplace=True)
# df_athelete_events_BMI["Medal"].fillna("NoMedal", inplace=True)
# df_athelete_events_BMI["Height"].fillna(value=round(df_athelete_events["Height"].mean()), inplace=True)
# df_athelete_events_BMI["Weight"].fillna(value=round(df_athelete_events["Weight"].mean()), inplace=True)
#------------------------------------------------------
df_athelete_events_BMI['BMI'] = df_athelete_events['Weight'] / ((df_athelete_events['Height']/100)**2)
```

```python
tt=df_athelete_events_BMI.Sport.values
cc= np.logical_or(tt=="Basketball",tt=="Swimming")
cc= np.logical_or(cc,tt=="Football")
#--------------------------------------------------
df_athelete_events_BMI=df_athelete_events_BMI[cc] #DataSet with only Basketball & Football & Swimming
df_athelete_events_BMI_Nomedal=df_athelete_events_BMI.query("Medal=='NoMedal'") # above dataSet for all atheletes who didnot win.
df_athelete_events_BMI_Medal=df_athelete_events_BMI.query("Medal!='NoMedal'")  # above dataSet for all atheletes who won medals.
```

```python
f, ax = plt.subplots(figsize=(5,4))
sns.despine(f)
sns.kdeplot(df_athelete_events_BMI_Nomedal['BMI'], shade=True,color='b')
sns.kdeplot(df_athelete_events_BMI_Medal['BMI'], shade=True,color='r')
top_bar = mpatches.Patch(color='r', label='Winners')
bottom_bar = mpatches.Patch(color='b', label='Losers')
plt.legend(handles=[top_bar, bottom_bar])
plt.title('BMI Density distribution')
plt.show()
```
